# Clustering Crypto

```python
# Initial imports
import pandas as pd
import hvplot.pandas
import numpy as np
from path import Path
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

### Deliverable 1: Preprocessing the Data for PCA

```python
# Load the crypto_data.csv dataset.
file_path = "crypto_data.csv"
crypto_df = pd.read_csv(file_path)
crypto_df = crypto_df.set_index('Unnamed: 0')
crypto_df.index.name = None 
crypto_df.head(10)
```

```python
# Keep all the cryptocurrencies that are being traded.
crypto_df = crypto_df[crypto_df.IsTrading != False]
crypto_df.shape
```

```python
crypto_df.head(10)
```

```python
# Keep all the cryptocurrencies that have a working algorithm.
```

```python
# Remove the "IsTrading" column. 
new_crypto_df = crypto_df.drop(["IsTrading"], axis = 1)
new_crypto_df
```

```python
# Remove rows that have at least 1 null value.
edited_crypto_df = new_crypto_df.dropna()
edited_crypto_df
```

```python
# Keep the rows where coins are mined.
edited_new_crypto_df = edited_crypto_df[edited_crypto_df['TotalCoinsMined'] > 0]
edited_new_crypto_df.head(10)
```

```python
# Create a new DataFrame that holds only the cryptocurrencies names.
CoinName_df = edited_new_crypto_df['CoinName'].to_frame()
CoinName_df.head(5)
```

```python
# Drop the 'CoinName' column since it's not going to be used on the clustering algorithm.
edited_new_crypto_df.drop('CoinName', axis=1, inplace=True)
edited_new_crypto_df
```

```python
crypto_df = edited_new_crypto_df
crypto_df
```

```python
# Use get_dummies() to create variables for text features.
X = pd.get_dummies(crypto_df, columns=["Algorithm", "ProofType"])
X
```

```python
# Standardize the data with StandardScaler().
X_scaled = StandardScaler().fit_transform(X)
X_scaled[0:5]
```

### Deliverable 2: Reducing Data Dimensions Using PCA

```python
# Using PCA to reduce dimension to three principal components.

# Initialize PCA model
pca = PCA(n_components=3)

# Get two principal components for the iris data.
crypto_pca = pca.fit_transform(X_scaled)
crypto_pca
```

```python
# Create a DataFrame with the three principal components.
pcs_X_df = pd.DataFrame(
    data=crypto_pca)
pcs_df = pd.DataFrame(pca.fit_transform(pcs_X_df), index=X.index, columns=['PC 1', 'PC 2', 'PC 3'])
pcs_df
```

### Deliverable 3: Clustering Crytocurrencies Using K-Means

#### Finding the Best Value for `k` Using the Elbow Curve

```python
# Create an elbow curve to find the best value for K.
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of K values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(pcs_df)
    inertia.append(km.inertia_)

# Create the elbow curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
#df_elbow.hvplot.line(x="k", y="inertia", xticks=k, title="Elbow Curve")
plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(list(range(11)))
plt.title('Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

Running K-Means with `k=4`

```python
# Initialize the K-Means model.
model = KMeans(n_clusters=4, random_state=1)

# Fit the model
model.fit(pcs_df)

# Predict clusters
predictions = model.predict(pcs_df)
predictions
```
